```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
data=iris.data
```

```python
y=iris.target
```

```python
from keras.utils import to_categorical
```

```python
targrt=to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
               data, targrt, test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
sX=scaler.transform(X_train)
```

```python
sx_test=scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model= Sequential()
```

```python
model.add(Dense(4,input_dim=4,activation='relu'))
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(sX,y_train,epochs=500,verbose=2)
```

```python
pre=model.predict_classes(sx_test)
```

```python
test_y=y_test.argmax(axis=1)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(pre,test_y))
print(accuracy_score(pre,test_y))
print(confusion_matrix(pre,test_y))
```
